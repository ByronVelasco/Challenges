## **First approach**

```python
def largest_product_in_series(series, k):
  max_product = 0
  for i in range(len(series)-k+1):
    product = 1
    for j in range(i, i+k):
      product *= int(series[j])
    if product > max_product:
      max_product = product
  return max_product
```

## **Optimal method**

```python
def largest_product_in_series_optimal(series, k):
  max_product = 0
  i = 0
  while i <= len(series)-k:
    prod_zero = False
    product = 1
    for j in range(i, i+k):
      if series[j] == '0':
        prod_zero = True
        break
      product *= int(series[j])
    if not prod_zero:
      i += 1
      if product > max_product:
        max_product = product
    else:
      i += k
  return max_product
```
